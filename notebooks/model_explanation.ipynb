```python
import joblib
import pandas as pd
import shap
```

```python
creditcard_random_forest = joblib.load("../models/creditcard_random_forest.pkl")
frauddata_random_forest = joblib.load("../models/frauddata_random_forest.pkl")
creditcard_logistic_regression = joblib.load("../models/creditcard_logistic_regression.pkl")
frauddata_logistic_regression = joblib.load("../models/frauddata_logistic_regression.pkl")
```

creditcard

```python
X = pd.read_csv("../Data/creditcard.csv")
y = X["Class"]
X = X.drop(columns=["Class"])
```

```python
X_sampled = X.iloc[:500]
```

```python
X_sampled.shape
```

```python
print(f"Explaining model: {creditcard_logistic_regression.__class__.__name__}")
explainer = shap.LinearExplainer(creditcard_logistic_regression, X_sampled)
shap_values = explainer.shap_values(X_sampled)
shap.summary_plot(shap_values, X_sampled)
```
